```python
def revenue(data, value):
    dict1 = {}
    
    for i in range(len(data)):
        month = data[i][5:7]
    
        if month not in dict1:
            dict1[month] = value[i]
        
        else:
            dict1[month] += value[i]
           
    return dict1



revenue(data, value)
```
